```python
import nltk
import string
from matplotlib import pyplot as plt 

stopwords = nltk.corpus.stopwords.words('turkish')
#stopwords.extend(["da", "ki", "de"])

# öznitelikler
corner_posts_by_author_tokens = {}
corner_posts_by_author_sent_tokens = {}
corner_posts_by_author_token_length_distributions = {}
corner_posts_by_author_sent_tokens_length_distributions = {}
for author in authors:
    # cümle tokenleri
    corner_posts_by_author_sent_tokens[author] = nltk.sent_tokenize(corner_posts_by_author[author])
    
    #kelime tokenleri
    tokens = []
    for sent in corner_posts_by_author_sent_tokens[author]:
        # noktalama işaretlerini filtreleme & büyük harfleri küçük harfe çevirme
        sent = "".join([i for i in sent.lower() if (i.isalpha() or i.isspace() or i.isnumeric())])
        tokens += nltk.word_tokenize(sent)
    
    # stopword - türkçede sık kullanılan kelimeleri çıkartalım
    corner_posts_by_author_tokens[author] = ([token for token in tokens if (token not in stopwords)])
    
    # kelime token uzunluklarının dağılımlarını alma
    token_lengths = [len(token) for token in corner_posts_by_author_tokens[author]]
    
    # cümle token uzunluklarının dağıımlarını alma
    sent_token_lengths = [len(nltk.word_tokenize(sent)) for sent in corner_posts_by_author_sent_tokens[author]]

    corner_posts_by_author_sent_tokens_length_distributions[author] = nltk.FreqDist(sent_token_lengths)
    corner_posts_by_author_token_length_distributions[author] = nltk.FreqDist(token_lengths)
    #corner_posts_by_author_sent_tokens_length_distributions[author].plot(15, title = author)
    #corner_posts_by_author_token_length_distributions[author].plot(15, title = author)
```

```python
#yazarların kelime ki-kare hesaplaması  
authors = ('Güneri Civaoğlu', 'Oray Eğin', 'Fatih Altaylı', 'Orhan Karataş', 'Didem Özel Tümer',
           'Rauf Tamer', 'Oral Çalışlar', 'Mehmet Tezkan', 'Kübra Par', 'Murat Yetkin') 
for author in authors:
    joint_corpus = (corner_posts_by_author_tokens[author] + corner_posts_by_author_tokens['Unknown'])
    joint_freq_dist = nltk.FreqDist(joint_corpus)
    most_common = list(joint_freq_dist.most_common(100))
    
    # yazar / yazar + bilinmeyen oranı
    author_share = (len(corner_posts_by_author_tokens[author]) / len(joint_corpus))
    
    #en yaygın kullanılan 100 kelimenin yazarlardaki dağılımları
    cisquared = 0
    for word, joint_count in most_common:
        #bu yaygın kelimeleri ne kadar sık görüyoruz?
        author_count = corner_posts_by_author_tokens[author].count(word)
        unknown_count = corner_posts_by_author_tokens['Unknown'].count(word)
        
        #ne sıklıkla görmeyi bekliyoruz?
        expected_author_count = joint_count * author_share
        expected_unknown_count = joint_count * (1-author_share)
        
        #ki-kareler hesaplanır
        cisquared += ((author_count - expected_author_count)**2 / expected_author_count)
        cisquared += ((unknown_count - expected_unknown_count)**2 / expected_unknown_count)
        
    print(author, "ki-kare istatistiği", cisquared)
```

```python
import math

#z-score hesaplama
whole_corpus =  []
for author in authors:
    whole_corpus += corner_posts_by_author_tokens[author]

# frekans dağılımı
whole_corpus_freq_dist = list(nltk.FreqDist(whole_corpus).most_common(30))

# temel veri yapısı
features = [word for word, freq in whole_corpus_freq_dist]

feature_freqs = {}
for author in authors:
    # yazarın özelliklerini tutan bir sözlük
    feature_freqs[author] = {}
    # yazarların token sayıları
    overall = len(corner_posts_by_author_tokens[author])
    for feature in features:
        presence = corner_posts_by_author_tokens[author].count(feature)
        feature_freqs[author][feature] = presence / overall
        
# külliyat standartları
corpus_features = {}

for feature in features:
    # özelliğin standart sapması ve ortalamasını tutacak sözlük
    corpus_features[feature] = {}
    # özelliğin ortalaması
    feature_average = 0
    for author in authors:
        feature_average += feature_freqs[author][feature]
    feature_average /= len(authors)
    corpus_features[feature]["Mean"] = feature_average
    
    # özelliğin standart sapması
    feature_stdev = 0
    for author in authors:
        diff = feature_freqs[author][feature] - corpus_features[feature]["Mean"]
        feature_stdev += diff**2
    feature_stdev /= (len(authors) - 1)
    feature_stdev = math.sqrt(feature_stdev)
    corpus_features[feature]["StdDev"] = feature_stdev
    
feature_zscores = {}
for author in authors:
    feature_zscores[author] = {}
    for feature in features:
        #Z-score = (value - mean) / stddev
        feature_val = feature_freqs[author][feature]
        feature_mean = corpus_features[feature]["Mean"]
        feature_stdev = corpus_features[feature]["StdDev"]
        feature_zscores[author][feature] = ((feature_val - feature_mean) / feature_stdev)
    
# unknown için hesaplamalar
overall = len(corner_posts_by_author_tokens['Unknown'])
unknown_freqs = {}
for feature in features:
    presence = corner_posts_by_author_tokens['Unknown'].count(feature)
    unknown_freqs[feature] = presence / overall
    
unknown_zscores = {}
for feature in features:
    feature_val = unknown_freqs[feature]
    feature_mean = corpus_features[feature]["Mean"]
    feature_stdev = corpus_features[feature]["StdDev"]
    unknown_zscores[feature] = ((feature_val - feature_mean) / feature_stdev)
    print("Unknown ", feature, " özelliği z-score'u : ", unknown_zscores[feature])
```
