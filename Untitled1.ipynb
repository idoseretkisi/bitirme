```python
authors = ('Rauf Tamer', 'Oral Çalışlar', 'Mehmet Tezkan', 'Kübra Par', 'Murat Yetkin', 'Unknown',
               'Güneri Civaoğlu', 'Oray Eğin', 'Fatih Altaylı', 'Orhan Karataş', 'Didem Özel Tümer')

corner_posts_by_author = {}
for author in authors:
    with open(f'data/{author}.txt', "r", encoding='utf-8') as f:
        corner_posts_by_author[author] = f.read()
```

```python
import nltk
from matplotlib import pyplot as plt

stopwords = nltk.corpus.stopwords.words('turkish')
stopwords.extend(["da", "ki", "de"])

# öznitelikler
cpa_word_tokens = {} # kelime tokenleri
cpa_sent_tokens = {} # cümle tokenleri
cpa_wt_len_dist = {} # kelime tokenleri uzunluk dağılımı
cpa_st_len_dist = {} # cümle tokenleri uzunluk dağılımı
for author in authors:
    # cümle tokenleri
    cpa_sent_tokens[author] = nltk.sent_tokenize(corner_posts_by_author[author])

    # kelime tokenleri
    tokens = []
    for sent in cpa_sent_tokens[author]:
        # noktalama işaretlerini filtreleme & büyük harfleri küçük harfe çevirme
        sent = "".join([i for i in sent.lower() if (i.isalpha() or i.isspace() or i.isnumeric())])
        tokens += nltk.word_tokenize(sent)

    # stopword - türkçede sık kullanılan kelimeleri çıkartalım
    cpa_word_tokens[author] = ([token for token in tokens if (token not in stopwords)])

    # kelime token uzunluklarının dağılımlarını alma
    token_lengths = [len(token) for token in cpa_word_tokens[author]]

    # cümle token uzunluklarının dağıımlarını alma
    sent_token_lengths = [len(nltk.word_tokenize(sent)) for sent in cpa_sent_tokens[author]]

    cpa_st_len_dist[author] = nltk.FreqDist(sent_token_lengths)
    cpa_wt_len_dist[author] = nltk.FreqDist(token_lengths)
    #cpa_st_len_dist[author].plot(15, title=author)
    #cpa_wt_len_dist[author].plot(15, title=author)
```

```python
from nltk.util import ngrams

cpa_quadgram = {} # 4gram
cpa_quadgram_dist = {} #4gram dağılımı
for author in authors:
    new_text = "".join([i for i in corner_posts_by_author[author].lower() if (i.isalpha() or i.isspace() or i.isnumeric())])
    cpa_quadgram[author] = list(ngrams(new_text.split(), n=4))
    cpa_quadgram_dist[author] = nltk.FreqDist(cpa_quadgram[author])
    #cpa_quadgram_dist[author].plot(10, title = author)
```

```python
cpa_avg_st_len = {} # ortalama cümle token uzunluğu
cpa_avg_wt_len = {} # ortalama kelime token uzunluğu
avg_st_len = 0
avg_wt_len = 0
for author in authors:
    sum_of_st_len = 0
    sum_of_wt_len = 0
    for feature in cpa_st_len_dist[author]:
        sum_of_st_len += feature
    avg_st_len = sum_of_st_len / len(cpa_st_len_dist[author])
    cpa_avg_st_len[author] = avg_st_len

    for feature in cpa_wt_len_dist[author]:
        sum_of_wt_len += feature
    avg_wt_len = sum_of_wt_len / len(cpa_st_len_dist[author])
    cpa_avg_wt_len[author] = avg_wt_len


fig, ax = plt.subplots()
ax.barh(list(cpa_avg_wt_len.keys()), list(cpa_avg_wt_len.values()))
ax.set(title='average word token')
plt.show()

fig2, ax2 = plt.subplots()
ax2.barh(list(cpa_avg_st_len.keys()), list(cpa_avg_st_len.values()))
ax2.set(title='average sent token')
plt.show()
```

```python
from snowballstemmer import TurkishStemmer
stemmer = TurkishStemmer()
cpa_stems = {} # stemler
cpa_stems_dist = {} # stem dağılımı

for author in authors:
    stems = [stemmer.stemWord(token) for token in cpa_word_tokens[author]]
    cpa_stems[author] = ([stem for stem in stems if (stem not in stopwords and len(stem) > 1)])
    cpa_stems_dist[author] = nltk.FreqDist(cpa_stems[author])
    #cpa_stems_dist.plot(15, title=author)
```

```python
#yazarların ki-kare hesaplaması

authors = ('Rauf Tamer', 'Oral Çalışlar', 'Mehmet Tezkan', 'Kübra Par', 'Murat Yetkin',
               'Güneri Civaoğlu', 'Oray Eğin', 'Fatih Altaylı', 'Orhan Karataş', 'Didem Özel Tümer')

cpa_wcq = {} #word cisquared
cpa_scq = {} #sent cisquared
cpa_stemcq = {} #stem cisquared
cpa_qgcq = {} #quadgram cisquared

features = [cpa_word_tokens, cpa_sent_tokens, cpa_stems, cpa_quadgram]
result = [cpa_wcq, cpa_scq, cpa_stemcq, cpa_qgcq]
for feature in range(0,4):
    for author in authors:
        joint_corpus = (features[feature][author] + features[feature]['Unknown'])
        joint_freq_dist = nltk.FreqDist(joint_corpus)
        most_common = list(joint_freq_dist.most_common(100))
    
        # yazar / yazar + bilinmeyen oranı
        author_share = (len(features[feature][author]) / len(joint_corpus))
    
        #en yaygın kullanılan 100 kelimenin yazarlardaki dağılımları
        cisquared = 0
        for word, joint_count in most_common:
        #bu yaygın kelimeleri ne kadar sık görüyoruz?
            author_count = features[feature][author].count(word)
            unknown_count = features[feature]['Unknown'].count(word)
        
            #ne sıklıkla görmeyi bekliyoruz?
            expected_author_count = joint_count * author_share
            expected_unknown_count = joint_count * (1-author_share)
        
            #ki-kareler hesaplanır
            cisquared += ((author_count - expected_author_count)**2 / expected_author_count)
            cisquared += ((unknown_count - expected_unknown_count)**2 / expected_unknown_count)
        
        result[feature][author] = cisquared
```
